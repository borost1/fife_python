```python
import matplotlib.pyplot as plt

abc = [
    'a', 'á', 'b', 'c', 'd', 'e', 'é', 'f', 'g', 'h', 'i', 'í', 'j', 'k', 'l', 'm', 'n',
    'o', 'ó', 'ö', 'ő', 'p', 'q', 'r', 's', 't', 'u', 'ú', 'ü', 'ű', 'v', 'w', 'x', 'y', 'z'
]

diph = ['cs', 'dz', 'gy', 'ly', 'ny', 'sz', 'ty', 'zs']
triph = 'dzs'
irasjel = [',', '.', '(', ')', '-']

szoveg_eredeti = "Külföldi jeladós barátkeselyűt lőhettek le Szabolcs-Szatmár-Bereg megyében, " \
         "írja a Magyar Madártani és Természetvédelmi Egyesület (MME) közleménye és a parlagisas.hu."
szoveg_eredeti = szoveg_eredeti.replace(' ', '')

szoveg = szoveg_eredeti
szoveg = szoveg.lower()

result = {}
result['_irasjel'] = 0
result['dzs'] = szoveg.count(triph)/len(szoveg_eredeti)*100
szoveg = szoveg.replace(triph, '')

for i in irasjel:
    result['_irasjel'] += szoveg.count(i)/len(szoveg_eredeti)*100
    szoveg = szoveg.replace(i, '')

for d in diph:
    result[d] = szoveg.count(d)/len(szoveg_eredeti)*100
    szoveg = szoveg.replace(d, '')

for d in abc:
    result[d] = szoveg.count(d)/len(szoveg_eredeti)*100
    szoveg = szoveg.replace(d, '')

result = dict(sorted(result.items()))  
print(result)
    
xs = result.keys()
ys = result.values()
    
plt.bar(xs, ys)
plt.rcParams["figure.figsize"] = (20,10)
plt.xlabel('betű')
plt.ylabel('%')
```
